```python
#Seasons for the data: slower in winter?

#spring = 3 - 5
#summer = 6 - 8
#fall = 9 - 11
#winter = 12 - 2

#split between 2 decimals
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3.head()
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Sping' if x >2 else 'Winter')
```

```python
df3.head(100)
```

```python
df3.groupby('race_season')['average_speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```
